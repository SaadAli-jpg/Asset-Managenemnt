```python
import csv
import os

FILE = 'assets.csv'

# Ensure the file exists
if not os.path.exists(FILE):
    with open(FILE, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['ID', 'Name', 'Type', 'Value', 'Location'])

def read_assets():
    with open(FILE, 'r', newline='') as f:
        reader = csv.DictReader(f)
        return list(reader)

def write_assets(assets):
    with open(FILE, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['ID', 'Name', 'Type', 'Value', 'Location'])
        writer.writeheader()
        writer.writerows(assets)

def add_asset():
    assets = read_assets()
    asset_id = str(int(assets[-1]['ID']) + 1 if assets else 1)
    name = input("Asset Name: ")
    type_ = input("Asset Type: ")
    value = input("Asset Value: ")
    location = input("Asset Location: ")
    assets.append({'ID': asset_id, 'Name': name, 'Type': type_, 'Value': value, 'Location': location})
    write_assets(assets)
    print("Asset added!\n")

def view_assets():
    assets = read_assets()
    print("\nID | Name | Type | Value | Location")
    print("------------------------------------")
    for a in assets:
        print(f"{a['ID']} | {a['Name']} | {a['Type']} | {a['Value']} | {a['Location']}")
    print()

def update_asset():
    assets = read_assets()
    asset_id = input("Enter Asset ID to update: ")
    for a in assets:
        if a['ID'] == asset_id:
            a['Name'] = input("New Name: ")
            a['Type'] = input("New Type: ")
            a['Value'] = input("New Value: ")
            a['Location'] = input("New Location: ")
            write_assets(assets)
            print("Asset updated!\n")
            return
    print("Asset ID not found!\n")

def delete_asset():
    assets = read_assets()
    asset_id = input("Enter Asset ID to delete: ")
    assets = [a for a in assets if a['ID'] != asset_id]
    write_assets(assets)
    print("Asset deleted!\n")

# Main loop
while True:
    print("1. Add Asset")
    print("2. View Assets")
    print("3. Update Asset")
    print("4. Delete Asset")
    print("5. Exit")
    choice = input("Choose option: ")

    if choice == '1':
        add_asset()
    elif choice == '2':
        view_assets()
    elif choice == '3':
        update_asset()
    elif choice == '4':
        delete_asset()
    elif choice == '5':
        break
    else:
        print("Invalid choice!\n")
```
